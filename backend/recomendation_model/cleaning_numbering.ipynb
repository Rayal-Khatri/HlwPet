```python
import pandas as pd

itemset = pd.read_csv('data\labelsv2.csv')
itemset
```

```python
import re 

def clean_title(title):
    return re.sub("[^a-zA-Z ]", "", title)
```

```python
itemset['breed'] = itemset["breed"].apply(clean_title)
itemset
```

```python
itemset.to_csv('breed_list.csv', index=False)
itemset
```
